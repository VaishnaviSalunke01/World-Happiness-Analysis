```python
df=df.dropna()
```

```python
df=df.rename(columns={
    'happiness_score': 'Happiness',
    'gdp_per_capita  ': 'GDP',
    'social_support': 'Social',
    'healthy_life_expectancy': 'Life_Expectancy',
    'freedom_to_make_life_choices': 'Freedom',
    'perceptions_of_corruption': 'Corruption'
    
})
```

```python
df.isnull().sum()
```

```python
##Visualization
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set_style('darkgrid')
```

```python
columns_to_correlate=['Happiness','gdp_per_capita','Social','Life_Expectancy','Freedom','generosity','Corruption']
correlation_matrix=df[columns_to_correlate].corr()

#draw heatmap
plt.figure(figsize=(10,7))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix of Happiness Factors')
plt.show()
```

```python
#Question 1: What is the single biggest driver of happiness?
     #Look at the "Happiness" row. The number closest to +1.0 (bright red) 
     #is the strongest positive driver. The number closest to -1.0 (bright blue) is the strongest negative driver.
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='gdp_per_capita', y='Happiness')
plt.title('Happiness vs. GDP per Capita')
plt.show()
```

```python
#Question 2: Does money really buy happiness?
  #Does the line go up and up? (More money = more happiness).

#Does it go up and then flatten out? (This is a huge insight: "Money buys happiness, but only up to a certain point!").

#Are there outliers? (e.g., A rich, sad country or a poor, happy country).
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Happiness'], kde=True, bins=20)
plt.title('Distribution of Global Happiness Scores')
plt.xlabel('Happiness Score')
plt.show()

#Q3.How is happiness distributed?
```

```python
#Question 4: The "Difficult Question" (Your Unique Insight)
  #This is what makes your project impressive. "What is the biggest driver of happiness in rich countries vs. poor countries?"

# 1. Find the middle-point for GDP
median_gdp = df['gdp_per_capita'].median()
print(f"The median GDP is: {median_gdp}")

# 2. Create two NEW dataframes: one for "low" GDP, one for "high"
low_gdp_df = df[df['gdp_per_capita'] < median_gdp]
high_gdp_df = df[df['gdp_per_capita'] >= median_gdp]

# 3. Now, just re-run the heatmap code from Question 1 on the first new dataframe
print("\n--- Correlation for LOW GDP Countries ---")
low_gdp_corr = low_gdp_df[columns_to_correlate].corr()
plt.figure(figsize=(10, 7))
sns.heatmap(low_gdp_corr, annot=True, cmap='coolwarm')
plt.title('Drivers of Happiness (LOW GDP Countries)')
plt.show()

# 4. ...and run it again on the second new dataframe
print("\n--- Correlation for HIGH GDP Countries ---")
high_gdp_corr = high_gdp_df[columns_to_correlate].corr()
plt.figure(figsize=(10, 7))
sns.heatmap(high_gdp_corr, annot=True, cmap='coolwarm')
plt.title('Drivers of Happiness (HIGH GDP Countries)')
plt.show()
```

```python
# **Finding:** The heatmap clearly shows that 'GDP' and 'Social' have the highest correlation with 'Happiness'
#(around +0.7 or +0.8). This means that, in general, richer and more supportive societies are happier.
```
